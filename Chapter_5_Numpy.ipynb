Chapter 5 - Numpy - Boolean Indexing

```python
import numpy as np
```

```python
names = np.array (['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
names
```

```python
data = np.random.randn (7,4)
data
```

```python
names == 'Bob' #to check if names of row equals to Bob
```

```python
data[names=='Bob']  #akan keluarkan row Bob only
```

```python
data[names=='Bob',2:]  #to call out column 2 onwards
```

```python
names != 'Bob'   #check everything except Bob
```

```python
data[~(names=='Bob')]   #select everything except Bob
```

```python
#to replace value
data[names=='Bob',3]=1
data
```

```python
mask = (names == 'Bob') | (names == 'Will')
data[mask]
#Note: To combine multiple Boolean conditions,
#use arithmetic: & (for and) and | (for or)
```

```python
data[data<0]=0   #to replace everything that is less than 0 = 0
data
```

```python
arr=np.arange(15)  #15 INDX
arr
```

```python
arr=np.arange(15).reshape((3,5))  #3 row, 5 column
arr
```

```python
arr.T   # 5x3
```

```python
np.dot(arr.T , arr)
#arr.T x arr  = (5x3)  x  (3x5)  =>  5x5
#make sure tengah tu sama number
# result = 1st row x 1st column, 2nd row x 2nd column
```

```python
np.dot(arr , arr.T)
#arr.T x arr  = (3x5)  x  (5x3)  =>  3x3
#make sure tengah tu sama number
# result = 1st row x 1st column, 2nd row x 2nd column
```
