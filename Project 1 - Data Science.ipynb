## PART 2 – LISTS, DICTIONARIES AND LOOPING

```python
# Create a List & Dictionary
cities= {
    'Nagoya' : {
         'Country' : 'Japan',
         'Population' : 123,
         'Landmarks': ['Nagoya Castle', 'Tokugawa Garden', 'Toyota Automobile Museum'],
         'Dishes' : ['Hitsumabushi', 'Tebasaki','Kishimen Udon'],
         },
    'Busan': {
         'Country' : 'South Korea',
         'Population' : 52,
         'Landmarks': ['Haeundae Beach', 'Gamcheon Culture Village', 'Busan Tower'] ,
         'Dishes' : ['Milmyeon', 'Eomuk', 'Jogae-gui'],
         },
    'Danang': {
         'Country' : 'Vietnam',
         'Population' : 105,
         'Landmarks': ['Dragon Bridge','Ba Na Hills', 'Marble Mountains'],
         'Dishes' : ['Mi Quang', 'Chicken Rice', 'Bun Cha Ca'],
         },
    'Bandung': {
         'Country' : 'Indonesia',
         'Population' : 281,
         'Landmarks': ['Kawah Putih', 'Tangkuban Perahu', 'Braga Street'],
         'Dishes' : ['Batagor', 'Nasi Timbel', 'Cireng'],
         },
    'Taipei': {
         'Country' : 'Taiwan',
         'Population' : 23,
         'Landmarks': ['Taipei 101', 'Shilin Night Market', 'Jiufen'],
         'Dishes' : ['Beef Noodle Soup', 'Bubble Tea', 'Scalion Pancake'],
         }
}


```

```python
# Use Loops to print name, country, population, landmarks & dishes
print("Cities that I would like to visit:")
for city, info in cities.items():
    country = info['Country'].title()
    population = info['Population']

    print(f"\nCity : {city.title()}")
    print(f"Country : {country}")
    print(f"Population: {population} million")

    print("3 Famous Landmarks:")
    for x in info["Landmarks"]:
        print(f"- {x}")

    print("3 Popular Local Dishes:")
    for x in info["Dishes"]:
        print(f"- {x}")
```
